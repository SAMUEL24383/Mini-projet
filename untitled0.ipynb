```python
print("Bienvenue dans le jeu de devinette")
import random
nombre_secret = random.randint(1, 100)
tentatives = 0
print("je pense à un nombre entre 1 et 100")

while True:
  proposition = int(input("Entrer votre estimation :"))
  tentatives += 1
  if proposition < nombre_secret:
    print("votre proposition est trop basse. Essayez encore !")
  elif proposition > nombre_secret:
    print("votre proposition est trop haute. Essayez encore !")
  else:
    print(f"Félicitations ! Vous avez deviné le nombre en {tentatives} tentatives.")
    break
```
