```python
chaîne_de_mot = input("Entrer une chaîne de la forme mot1/mot2/mot3/mot4/mot5/mot6/mot7")
liste_de_mots = chaîne_de_mot.split("/")
print(liste_de_mots)
from random import shuffle
shuffle(liste_de_mots)
print(liste_de_mots)
# Corrected list slicing and if-else logic
if len(liste_de_mots) < 10:
  # To print the first two elements, use slicing [0:2]
  print(liste_de_mots[0:2])
else:
  # This 'else' covers the case where len(liste_de_mots) >= 10
  # To print elements at indices 4, 5, 6, use slicing [4:7]
  print(liste_de_mots[4:7])
```
